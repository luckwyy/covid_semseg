```python


import numpy as np
import matplotlib.pyplot as plt
 
from sklearn.metrics import auc
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
 
```

```python
densenet_path = 'F:/tprfpr/densenet121_.csv'
googlenet_path = 'F:/tprfpr/googlenet_.csv'
resnet18_path = 'F:/tprfpr/resnet18_.csv'
resnet50_path = 'F:/tprfpr/resnet50_.csv'
resnet101_path = 'F:/tprfpr/resnet101_.csv'
```

```python
densenet_df = pd.DataFrame(pd.read_csv(densenet_path, low_memory=False))
googlenet_df = pd.DataFrame(pd.read_csv(googlenet_path, low_memory=False))
resnet18_df = pd.DataFrame(pd.read_csv(resnet18_path, low_memory=False))
resnet50_df = pd.DataFrame(pd.read_csv(resnet50_path, low_memory=False))
resnet101_df = pd.DataFrame(pd.read_csv(resnet101_path, low_memory=False))
```

```python
resnet101_df.head()
```

```python
 
densenet121_tpr = densenet_df['densenet121_bl_tpr']
densenet121_fpr = densenet_df['densenet121_bl_fpr']
googlenet_tpr = googlenet_df['googlenet_bl_tpr']
googlenet_fpr = googlenet_df['googlenet_bl_fpr']
resnet18_tpr = resnet18_df['resnet18_bl_tpr']
resnet18_fpr = resnet18_df['resnet18_bl_fpr']
resnet50_tpr = resnet50_df['resnet50_bl_tpr']
resnet50_fpr = resnet50_df['resnet50_bl_fpr']
resnet101_tpr = resnet101_df['resnet101_bl_tpr']
resnet101_fpr = resnet101_df['resnet101_bl_fpr']
# 设置终点t是1,
densenet121_tpr[0] = 1
googlenet_tpr[0] = 1
resnet18_tpr[0] = 1
resnet101_tpr[0] = 1
resnet50_tpr[0] = 1

densenet121_fpr[0] = 1
googlenet_fpr[0] = 1
resnet18_fpr[0] = 1
resnet101_fpr[0] = 1
resnet50_fpr[0] = 1
 
densetnet121_roc_auc = auc(densenet121_fpr, densenet121_tpr) ###计算auc的值
googlenet_roc_auc = auc(googlenet_fpr, googlenet_tpr) ###计算auc的值
resnet18_roc_auc = auc(resnet18_fpr, resnet18_tpr) ###计算auc的值
resnet50_roc_auc = auc(resnet50_fpr, resnet50_tpr) ###计算auc的值
resnet101_roc_auc = auc(resnet101_fpr, resnet101_tpr) ###计算auc的值
 
print(densetnet121_roc_auc)
 
plt.figure()
lw = 2
plt.figure(figsize=(15,10))

plt.plot(densenet121_fpr, densenet121_tpr, color='darkorange', lw=lw, label='densenet121 bl ROC curve (area = %0.5f)' % densetnet121_roc_auc) 

plt.plot(googlenet_fpr, googlenet_tpr, color='darkgoldenrod', lw=lw, label='googlenet bl ROC curve (area = %0.5f)' % googlenet_roc_auc) 

plt.plot(resnet18_fpr, resnet18_tpr, color='deeppink', lw=lw, label='resnet18 bl ROC curve (area = %0.5f)' % resnet18_roc_auc) 

plt.plot(resnet50_fpr, resnet50_tpr, color='green', lw=lw, label='resnet50 bl ROC curve (area = %0.5f)' % resnet50_roc_auc) 

plt.plot(resnet101_fpr, resnet101_tpr, color='red', lw=lw, label='resnet101 bl ROC curve (area = %0.5f)' % resnet101_roc_auc) 

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
 
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('bl')
plt.legend(loc="lower right")
plt.show()
```
